# 5. SORU - Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

```python
#öncelikle nobel tablosunun kopyası oluşturuldu. birthdate sütunundan yıl sütununu çıkarmamız gerekiyor çünkü kazanma yaşının sütununu yapacağız ve bu sayılara göre de bir iç göreü çıkaracağız.
#birthdate veri tipi stringten birthdate türüne çevrildi ve çıkarma işlemi yapıldı.

nobel1 = nobel.copy()
nobel1['birth_date'] = pd.to_datetime(nobel1['birth_date'], errors='coerce')
nobel1['winner_age'] = nobel1['year'] - nobel1['birth_date'].dt.year

nobel1
```

```python
#yukarıda ki tabloda ki veri lerin x eksenine kazanım yılı ve y eksenine kazanma yaşını koyarak scatter plot grafiği yapıldı.
#Erkek ve Kadın kazananlarda iki ayrı satırda gösterildi.
#kategoriler sütun halinde gösterildi.
#yazım sıralaması da belirtildi.
#ve başlık eklendi.

fig2 = px.scatter(nobel1, x="year", y='winner_age', color='sex', height=700, width=3000,
                  facet_row='sex', facet_col="category",
                  category_orders={"sex": ["Male", "Female"],
                                   "category": ["Chemistry", "Literature", "Medicine", "Peace", "Physics", "Economics"]},
                  labels={"sex": "Cinsiyet", "category": "Kategori", "year": "Yıl", "winner_age": "Kazanan Yaş"},
                  title="Nobel Ödül Kazananların Yaşları ve Kategorilerine Göre Dağılımı")

fig2.update_layout(
    title=dict(text="Nobel Ödül Kazananların Yaşları ve Kategorilerine Göre Dağılımı", font=dict(size=24)),
)

fig2.show()
```

```python
#en başta çıkardığımız tablodan yine yeni bir tablo yaratarak sütun silindi. axis = 1 yazılması gerekiyor çünkü yukarıdan aşağı silinmesi gerektiğinin komutu yazılmalıdır.

nobel2 = nobel.drop(["prize","birth_date","sex","full_name"], axis = 1)

nobel2
```

# 7. SORU - 1947-1991 yılları arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.(Her kategori için ayrı bir grafik olması istenmektedir)



```python
#filtreleme işlemi yapıldı.

nobel2_filtered_47_91 = nobel2[(nobel2['year'] >= 1947) & (nobel2['year'] <= 1991)].reset_index(drop=True)

nobel2_filtered_47_91
```

```python
#en çok kazanan ülke verisinde bu tarih aralığında yüksekten küçüğe toplandı ve tablo yazdırıldı.

nobel2_filtered_47_91_grouped = nobel2_filtered_47_91.groupby('birth_country').size().reset_index(name='Toplam').nlargest(10, 'Toplam')
nobel2_filtered_47_91_grouped
```

```python
# 6 kategoriyi de farklı gruplandırark grafiklerde karşılaştırmak amacıyla en iyi görseli treemap yapacağı düşünüldü.
# Kategorilerin hepsi farklı değerlendirildi ve kendi içlerinde toplamları gösterildi.

nobel2_filtered_47_91_grouped = nobel2_filtered_47_91.groupby(['category', 'birth_country']).size().reset_index(name='count')


fig5 = px.treemap(nobel2_filtered_47_91_grouped, path=['category', 'birth_country'], title="ÜLKE BAZINDA KAZANAN KATEGORİLER 1947-1991",
                  values='count',
                  color='count',
                  hover_data=['birth_country'],
                  color_continuous_scale='Mrybm')

fig5.update_layout(margin=dict(t=65, l=45, r=25, b=45))
fig5.update_layout(title_x=0.5)
fig5.update_layout(
    title=dict(text="ÜLKE BAZINDA KAZANAN KATEGORİLER 1947-1991", font=dict(size=24)),
)
fig5.show()
```

# 8. SORU - Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.

```python
nobel1
```

```python
#gereksiz kolonlar tekrar silindi.

nobel3 = nobel1.drop(["prize","full_name","birth_date","sex"], axis = 1)

nobel3
```

```python
#5 kategori üzerinde filtreleme işlemi yapıldı.

nobel3_filtered = nobel3[(nobel3['year'] >= 2000) & ((nobel3['category'] == "Chemistry") |
                                                      (nobel3['category'] == "Literature") |
                                                        (nobel3['category'] == "Peace") |
                                                         (nobel3['category'] == "Medicine") |
                                                          (nobel3['category'] == "Physics"))].reset_index(drop=True)

nobel3_filtered
```

```python
#yaşlar üzerinde bir iç görü çıkarmak için tablo açıklaması çıkarıldı.

nobel3_filtered.describe()
```

```python
#ekstra grouplama yapıp önce kategori sonra doğum ülkesi ve yaşı üzerinde gruplandırıldı. Fakat bu veri görselleştirilmedi.
nobel3_filtered_grouped = nobel3_filtered.groupby(['category', 'birth_country','winner_age']).size().reset_index(name='count')

nobel3_filtered_grouped
```

```python
nobel3_filtered_grouped = nobel3_filtered.groupby(['category', 'birth_country']).agg({'winner_age': 'mean'}).reset_index()

fig11 = px.bar(nobel3_filtered_grouped, x='birth_country', y='winner_age', color='category', height=1200, width=3000,
             facet_row='category',
             category_orders={"category": ["Chemistry", "Literature", "Peace", "Physics", "Medicine"]},
             labels={'winner_age': 'Yaş', 'birth_country': 'Doğum Ülkesi'},
             title='2000 Sonrasındaki Nobel Kazananlarının Ülkeleri ve Yaşları')
fig11.update_layout(
    title=dict(text='2000 Sonrasındaki Nobel Ödülü Kazananlarının Ülkeleri ve Yaşları', font=dict(size=24)),
)
fig11.show()
```

Kazanan ödül sahiplerinin de doğum ve yüksek öğretime kadar olan zamanı 2. dünya savaşı sonrası gerçekleşen büyük değişim içinde büyütülen ve eğitim gören düşünür ve öncüler olduğu görülüyor.

## Bütün ödül almış ülkelerde ilginç bir patern olarak görülecek bir durumda: bilim dalları olan kimya, fizik ve tıp alanlarında Dünya gayri safi milli hasılasının yüksek olan ve yatırım için kuvvetli olan ülkeleri (Avrupa ülkeleri, Kuzey Amerika, Uzak doğu) bulunuyor. Edebiyat ve Barış ödülleri bu kuvveti düşük olan veya hiç kuramayan ülkelerde bulunuyor (Ortadoğu, Afrika ve Güney Amerika). Bu durumun sebebi politik sebepler olduğunu düşünebiliriz.

## Bütün dallarda ödül almış sadece 2 ülke gözüküyor. Bunlar Çin ve ABD olarak gözüküyor. Ülkelerin değerlendirmelerini yaparken belirtildiği gibi ticaret ve üretim kanallarında yapılan değişim ve yapılanmaların başlangıcı olabilir.  



# ---------  GENEL DEĞERLENDİRME  ------------


## Nobel ödüllerinin her yıl farklı ülkelere gitmesinin ekonomik, sosyal, çevresel, politik, teknolojik ve refah seviyesi yüksek veya düşük olmasına direk bağlantılı olduğunu düşünmemenin mümkün olmadığı söylenebilir. Avrupa ülkelerinde yaşanan büyük değişimlerin getirdiği etkiler, ekonomik seviyede ticaret kanallarının artması, birleşmiş milletlerin kurulması ve birlik bilinciniin artması, üretimin bilgi ve yeni enerji buluşlarıyla hızla yükselmesi, bilginin belirli ülkelerde diğerlerine göre daha kolay ulaşılabilirliği, yatırım imkanları ve gelişim bilinci ülkeler arasında ki farkılılıkları bu ödüllerin sonuç analizinde de izlenebiliyor.
